```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# arange: used to create an array
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
# g indicates green (c means color)
plt.scatter(x,y,c='g')
```

```python
# labelling
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

```python
plt.scatter(x,y,c='b')
# labelling
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graphs in 2D')
# to save
plt.savefig('Test.png')
```

```python
# linewidth is width of the line and markersize is the size of dots
plt.plot(x,y,'ro',linestyle='dashed',linewidth=3,markersize='8')
```

```python
# sub-plots: within one canva(diagram) ploting multiple diagrams
# subplot(rows,columns,position)
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*-')
plt.subplot(2,2,3)
plt.plot(x,y,'bo-')
```

```python
# 3rd argu. is step-size
c= np.arange(0,4*np.pi,0.1)
d = np.sin(c)
plt.plot(c,d)
```

```python
# Compute the x and y coordinates for points on sine and cosine curves 
c = np.arange(0, 5 * np.pi, 0.1) 
d_sin = np.sin(c) 
d_cos = np.cos(c)
    
plt.subplot(2, 1, 1)    
plt.plot(c, d_sin,'r--') 
plt.title('Sine')  
     
plt.subplot(2, 1, 2) 
plt.plot(c, d_cos,'g--') 
plt.title('Cosine')  
   
plt.show()
```

```python
# Pie chart
```

```python
labels='python','c++','ruby','java'
sizes=[215,130,245,210]
colors=['gold','yellowgreen','lightcoral']
# explode 1st slice, autopct means in which format we want .
# here floating format
explode=(0.1,0,0.2,0)
# plot
# explode means how much far the slice will go.
plt.pie(sizes,explode=explode,labels=labels,colors=colors,
        autopct='%1.1f%%',shadow=True)
plt.axis('equal')
plt.show()
```
